### Dealing with missing datas 

```python
car_sales_missing = pd.read_csv("car-sales-missing-data.csv")
car_sales_missing
```

```python
# [fillna]
car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean())
```

```python
# [fillna with inplace] with inplace true we dont need to reassign the value back to the table as it will do it for us 
car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean(), inplace=True)
car_sales_missing
```

```python
# [dropna] remove row with na value 
car_sales_missing.dropna()
```

```python
car_sales_missing_dropped = car_sales_missing.dropna()
car_sales_missing_dropped
```

```python
car_sales_missing_dropped.to_csv("car-sales-missing-dropped.csv")
```

## Manipulating Data [Part2]
- create a new column with pd.Series
- create a new column with python approach
- create a new column with some calculation 
- [drop()] to delete a coolumn 

```python
# Colunm from series
seats_column = pd.Series([5, 5, 5, 5, 5])

# New colunm called seats 
car_sales["Seats"] = seats_column
car_sales
```

```python
car_sales["Seats"].fillna(5, inplace=True)
car_sales
```

```python
# Column from Python list (the lenght of the array has to be the same of the number of rows of the table)
fuel_economy = [7.5, 9.2, 5.0, 9.6, 8.7, 3.0, 7.6, 4.7, 4.5, 3.0]
car_sales["Fuel per 100KM"] = fuel_economy
car_sales
```

```python
car_sales["Total fuel used (L)"] = car_sales["Odometer (KM)"]/100 * car_sales["Fuel per 100KM"]
car_sales
```

```python
# Create a colunm from a single value 
car_sales["Number of wheels"] = 4
car_sales
```

```python
car_sales["Passed road safety"] = True
car_sales
```

```python
car_sales.dtypes
```

```python
# [drop] To drop a column (axis=1 == column; axis=0 == row)
car_sales = car_sales.drop("Total fuel used", axis=1)
car_sales
```

## Manipulating Data [Part3]
- sample is a great way to shuffle the dataframe to mix up the indexes (this will be use when training model to avoid bias)
- [reset_index()]
- [apply()] to use function inside of it such as "lamda"

```python
# frac is fraction 
car_sales.sample(frac=0.5)
```

```python
car_sales_shuffled = car_sales.sample(frac=1)
car_sales_shuffled
```

```python
# Only select 20% of the data
car_sales_shuffled.sample(frac=0.2)
```

```python
# [reset_index()] to reset the data
car_sales_shuffled.reset_index()
```

```python
# To remove the extra index field after reset_index we need to call drop param
car_sales_shuffled.reset_index(drop=True, inplace=True)
car_sales_shuffled
```

```python
car_sales
```

```python
# [apply] to use a function inside of it
# lambda is an anonymous function in Python
car_sales["Odometer (KM)"] = car_sales["Odometer (KM)"].apply(lambda x: x / 1.6)
car_sales
```
