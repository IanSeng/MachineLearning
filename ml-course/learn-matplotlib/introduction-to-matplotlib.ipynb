```python
# Make a plot from dictionary 
nut_butter_prices = {"Almond butter": 10,
                    "Peanut butter": 8,
                    "Cashew butter": 12};
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Dan's Nut Butter Store",
      ylabel="Prices ($)");
```

```python
# for barh the axis has to be list
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()),list(nut_butter_prices.values()));
```

### Two options for subplots

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10, 5))
# Plot to each different axis 
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplots options 2
fig, ax = plt.subplots(nrows = 2,
                      ncols = 2,
                      figsize=(10, 5))

# Plot to each different index
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

### Which one should you use?  (pyplot vs matplotlib OO method?)
* When plotting something quickly, okay to use pyplot method
* When plotting something more advance, use the OO method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
len(over_50)
```

```python
# Pyplot method (plotting directly from pandas data)
# We can see this method is not producing X axis label
over_50.plot(figsize=(10, 6), kind="scatter",
            x="age",
            y="chol",
            c="target"); # c is color
```

```python
# OO method mixed with pyplot method
fix, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            ax=ax);
ax.set_xlim([45, 100])
```

### Plotting from Pandas DataFrames 6 

```python
over_50["target"].values
```

```python
# OO method from scratch 
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"]);

# Customize the plot 
ax.set(title = "Heart Disease and Cholesterol Levels",
      xlabel="Age",
      ylabel="Cholesterol");

# Add a legend (1/0 label)
ax.legend(*scatter.legend_elements(), title="Targe");

# Add a horizontal line 
ax.axhline(over_50["chol"].mean(),
          linestyle="--");
```

### Plotting from Pandas DataFrames7

```python
# Subplot of chol, age, thalach 
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10, 10), sharex=True)

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"]);
        
# Customize ax0 
ax0.set(title="Heart Disease and Cholesterol Levels",
       ylabel="Cholesterol")

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

# Add a mean horizontal line 
ax0.axhline(over_50["chol"].mean(),
           linestyle="--")


# Add data to ax1
scatter1 = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"]);

# Add lebel
ax1.set(title="Heart Disease and Max Heart Rate (Thalach)",
       xlabel="Age",
       ylabel="Thalach");

# Add legend 
ax1.legend(*scatter.legend_elements(), title="Target");

# Add horizontal line 
ax1.axhline(over_50["thalach"].mean(),
           linestyle="--");

# Add title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold")
```

```python
## Set the style
plt.style.use('seaborn-whitegrid')

## OO method from scaratch
fig, ax = plt.subplots(figsize = (10, 6))

# Plot the data 
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"], 
                    cmap="winter") # this changes the colour scheme 

# Customize the plot 
ax.set(title="Heart Diease and Cholesterol Levels",
      xlabel="Age",
      ylabel="Cholesterol")

# Add a legend 
ax.legend(*scatter.legend_elements(), title="Target");

# Add a horizontal line 
ax.axhline(over_50["chol"].mean(), linestyle="--");
```

```python
# Customizing the y and x axis limitations

# Subplot of chol, age, thalach 
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10, 10), sharex=True)

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                     cmap="winter");
        
# Customize ax0 
ax0.set(title="Heart Disease and Cholesterol Levels",
       ylabel="Cholesterol")

# Change the x axis limit since share x is true it will overwrite the
# other subplot
ax0.set_xlim([50, 70])

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

# Add a mean horizontal line 
ax0.axhline(over_50["chol"].mean(),
           linestyle="--")


# Add data to ax1
scatter1 = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"],
                      cmap="winter");

# Add lebel
ax1.set(title="Heart Disease and Max Heart Rate (Thalach)",
       xlabel="Age",
       ylabel="Thalach");

# Add legend 
ax1.legend(*scatter.legend_elements(), title="Target");

# set y axis limit 
ax1.set_ylim([60, 200]);

# Add horizontal line 
ax1.axhline(over_50["thalach"].mean(),
           linestyle="--");

# Add title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
